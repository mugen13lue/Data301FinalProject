```
import pandas as pd
import numpy as np
import os
import json
from pandas.io.json import json_normalize

dir_main = "Data301FinalProject/datasets/"
!ls $dir_main
```

```
dir201902 = dir_main + "ks_2019-02-14/"
ks201902 = pd.read_csv(dir201902 + "Kickstarter000.csv",
                       engine="python")
ks201902 = ks201902.drop(
    ["is_backing", "is_starrable", "is_starred", "permissions",
     "currency_symbol", "friends", "source_url", "disable_communication",
     "currency_trailing_code", "current_currency", "usd_type", "pledged",
     "static_usd_rate", "fx_rate", "photo",
     "converted_pledged_amount", "profile"],
    axis=1
)

# TODO: weren't able to successfully convert creator and location to JSON.
# dropping these cols and moving on for now, but should revisit later
ks201902 = ks201902.drop(["location", "creator"], axis=1)
```

```
import ast

json_cols = ["category", "urls"]
#json_cols = ["category", "creator", "photo", "urls", "location"]
# TODO: weren't able to successfully convert creator and location to JSON.
# ignoring these cols and moving on for now, but should revisit later

for col in json_cols:
  print(col)
  ks201902[col] = ks201902[col].fillna('')
  ks201902[col] = ks201902[col].apply(ast.literal_eval)
  
ks201902.head(3)
```

## Clean JSON Variables

```
category_exp = json_normalize(ks201902["category"])
ks201902 = ks201902.drop("category", axis=1)
ks201902.head()
```

```
category_tups = category_exp.slug.str.split("/").apply(pd.Series)
category_tups.columns = ["parent_category", "category"]
ks201902 = pd.concat([ks201902, category_tups], axis=1)
ks201902.head()
```

```
urls_exp = json_normalize(ks201902["urls"])
urls_exp = urls_exp[["web.project", "web.rewards"]]
urls_exp.columns = ["project_url", "rewards_url"]
urls_exp.head()
```

```
ks201902 = pd.concat([ks201902, urls_exp], axis=1)
ks201902 = ks201902.drop("urls", axis=1)
ks201902.head()
```

## Clean Timestamps

```
from datetime import datetime

timestamps = ["created_at", "state_changed_at", "launched_at", "deadline"]
for ts in timestamps:
  ks201902[ts] = ks201902[ts].apply(datetime.utcfromtimestamp)
ks201902.head()
```

```
ks201902["project_duration"] = ks201902["deadline"] - ks201902["created_at"]
ks201902["time_til_state_changed"] = (ks201902["state_changed_at"] -
                                      ks201902["created_at"])
ks201902.head()
```

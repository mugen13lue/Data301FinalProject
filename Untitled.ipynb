```python
ks201901["avg_backer_plegde"] = ks201901['usd_pledged']/ks201901['backers_count']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import StandardScaler


Y_train = ks201901[ks201901["state"] != "live"]
Y_train = Y_train[Y_train["state"] != "canceled"]
Y_train = Y_train[Y_train["state"] != "suspended"]

X_test = ks201901[ks201901["state"] == "live"]
X_test = X_test[["avg_backer_plegde", "goal"]].fillna(0)

ks201901done = ks201901[(ks201901["state"] != "live")]
ks201901done = ks201901done[(ks201901done["state"] != "canceled")]
ks201901done = ks201901done[(ks201901done["state"] != "suspended")]

Y_train = Y_train["state"]
X_train = ks201901done[["avg_backer_plegde", "goal"]].fillna(0)
vec = DictVectorizer(sparse=False)
X_train_dict = vec.fit_transform(X_train.to_dict(orient="records"))

scaler = StandardScaler()
scaler.fit(X_train)
#keeping track of mean and standard deviation

X_train_sc = scaler.transform(X_train_dict)
X_test_sc = scaler.transform(X_test)

modelKnear = KNeighborsClassifier(n_neighbors=13)
modelKnear.fit(X_train_sc,Y_train)
quant_pred = modelKnear.predict_proba(X_test_sc)
modelKnear.predict(X_test_sc)
```

```python
#****** Cited Thomas's notebook.********
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt


def rfr(data, X,Y, Title):
    #data= dataframe that contains both input and output vars
    #X is an array of input column names
    #y is name of output column you are trying to predict
    #Title = Title of the graph
    
    tests = {
        "Number of Estimators" : [],
        "TestError":[],
    }
    
    for k in range(1, 75):
        model = RandomForestClassifier(n_estimators=k)
        test_error = cross_val_score(
                                 model,data[X].fillna(0), 
                                 data[Y],
                                 cv = 10, 
                                 scoring="accuracy"
                                ).mean()
        tests["Number of Estimators"].append(k)
        tests["TestError"].append(test_error)
        #print('.')
    df = pd.DataFrame(tests)
    ax = df.plot.scatter("Number of Estimators","TestError")
    ax.set_xlabel("Number of Estimators")
    ax.set_ylabel("Accuracy")
    fig = plt.gcf()
    fig.set_size_inches(16,6)
    plt.title(Title)
    print(df.loc[df.TestError.idxmax()])
    return df.sort_values("TestError").head(5)
```

So what we got what is really cool is that knearest neighbors gives us an accuracy of .9815 and random forest gives us an accuracy of .974 when we use "avg_backer_plegde", "goal", "usd_pledged", "backers_count" as features. This makes sense that average backer pledged will help predict if a project is successful because if the average backer is willing to pay a "good amount of money" then the likelihood that they will be successful will increase. Also using the goal as part of the feature will make sure things are scaled properly, so that projects requesting a little bit of money over projects ask for 1,000's of dollars can have similar average backer pledged, but have completely different states. if average backer pledged is high of course they will be successful, but if it is not high than what is the cutoff that the project will fail. certainly the cutoff for the project with a small goal will be a lot lower than the project asking for 1,000's of dollars. We also included usd pledged and backers count because they show the true value of total backers and pledged amounts. Yes, we know that the live projects will not have the same finished backers count or usd pledged, but this variable is a way to see how close they are to the goal.
